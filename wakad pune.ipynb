```python
from selenium import  webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd


chrome_options = webdriver.ChromeOptions()
### This blocks images and javascript requests
chrome_prefs = {
    "profile.default_content_setting_values": {
        "images": 2,
        "javascript": 2,
    }
}
chrome_options.experimental_options["prefs"] = chrome_prefs
###
driver=webdriver.Chrome(ChromeDriverManager().install())
# driver = webdriver.Chrome()

url = 'https://housing.com/in/buy/searches/P2kiztglzttku35yi'
driver.get(url)

time.sleep(2)
try:
    element = driver.find_element(By.TAG_NAME,value='body')
    while True:
        element.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
        page_41 = driver.find_element(By.CLASS_NAME,value='css-1nm99ps')


        m= '100'
        if m in page_41.text :
            break
except Exception as e:
    print('step1:',e)
    pass

try:
    
    bigbox = driver.find_elements_by_css_selector('.css-zrd0bv')
         
    PRICE=[]
    BHK=[]
    AREA=[]
    EMI=[]
    OWNER=[]
    list_of_links=[]
    RERA=[]
    for box in bigbox :
        #PRICE
        try: 
            PRICE.append(box.find_element_by_css_selector('.css-18rodr0').text)
        except:
            PRICE.append('NA')
         
        
        #BHK
        try:
            BHK.append(box.find_element_by_css_selector('.css-197fqpq').text)
        except:
            try:
                BHK.append(box.find_element_by_css_selector('.css-11nfaq3').text)
            except:
                BHK.append('NA')
#             BHK.append('NA')
            
        
        #AREA 
        try:
            AREA.append(box.find_element_by_css_selector('.css-4z3njv').text)
        except:
            AREA.append('NA')
        
        #EMI
        try:
            EMI.append(box.find_element_by_css_selector('.css-nin1gj').text)
        except:
            EMI.append('NA')
        
        #OWNER
        try:    
            OWNER.append(box.find_element_by_css_selector('.css-wni7av').text)
        except:
            OWNER.append('NA')
            
        #LINK
        try:
            x=box.find_element_by_tag_name('a')
            list_of_links.append(x.get_attribute('href'))
        except:
            list_of_links.append('NA')
        
        # RERA
        try:
            RERA.append(box.find_element_by_css_selector('.css-1lqxvi5').text)
        except:
            RERA.append('NOT RERA')
#         PRICE.append(price.text)
#         LAYOUT.append(BHK.text)
#         OWNER.append(owner.text)


#################new line added for saving the data##############
#         try:
#             df ={}
#             df['bhk']=BHK
#             df['price']=PRICE
#             df['emi']=EMI
#             df['sq_ft']=AREA
#             df['owners']=OWNER
#             df['link']=list_of_links
#             df['rera']=RERA
#         except Exception as e:
#             print('step3',e)
#             pass
        
# #         print(df)

#         try:
#             with open(f'wakad_pune_scraped.json',"r+",encoding="utf8") as f:
#                 data = json.load(f)
#                 data.update(df)
#                 f.seek(0)
#                 json.dump(data,f,ensure_ascii=False)
#                 print("SAVEDDDDDDDDDDDDD")
#         except:
#             print("COULDNT SAVE")
#             pass
# #             f= open('wakad_pune_scraped.json',"r+")
# #             data = json.load(f)
# #             print('daaaattttaaaaaaaaaaa',data)
# #             data.write(f.dump(df))
# # #             print('aaffttteerrrrrrrrrrr',data)
# # #             json.dump(data,f,ensure_ascii=False)
# # #             f.close()
# #             print("SAVEDDDDDDDDDDDDD")
# #             print('---------------------------------')
# #         except:
# #             print("COULDNT SAVE")
            
        
#######################################################################################        
          
except Exception as e:
    print('step2:',e)
    pass



try:
    df = pd.DataFrame()
    df['bhk']=BHK
    df['price']=PRICE
    df['emi']=EMI
    df['sq_ft']=AREA
    df['owners']=OWNER
    df['link']=list_of_links
    df['rera']=RERA
except Exception as e:
    print('step3',e)
    pass


df.to_csv('link_data_wakad_test.csv')
```
