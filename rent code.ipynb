```python
import pandas as pd
import requests

""" Import required libraries """

from bs4 import BeautifulSoup
from requests import get
import time
```

```python
list_data =[]
for i in range(1,5):
    
#     url = f'https://www.99acres.com/property-in-pune-ffid-page-'+str(i)
     
    url=f'https://housing.com/in/buy/pune/pune?page={i}'
    data = get(url)
    soup = BeautifulSoup(data.content, 'html.parser')

    housing_data = soup.find_all('div',attrs= {"class":"css-zrd0bv"})
    print(len(housing_data))
    list_data=[]
    for item in housing_data:

#         """ Verification ID"""
        try:
            price = item.find('div', attrs = {'class':'css-18rodr0'}).text
        except:
            price = 'NA'

#         BHK
        try:
            BHK =item.find('div', attrs = {"class":"css-197fqpq"}).text
#             print('price detect')
        except:
            try:
                BHK =item.find('div', attrs = {"class":"css-11nfaq3"}).text
            except:
                BHK = 'NA'

#         """AREA"""
        try :
            area= item.find('div', attrs = {"class":"css-hboir5"}).text
            
        except:
            area = 'NA'

# #         """owner"""
        try :
            owner = item.find('div', attrs = {"class":"css-wni7av"}).text 
            
        except:
            owner = 'NA'

#         """EMI"""
        try:
            EMI = item.find('span', attrs={"class": "css-nin1gj"}).text
        except:
            EMI = 'NA'

#         """LINK"""

        try:
            x=item.find_all('a',href=True)[0]
            link =x['href']
        except:
            link = 'NA'


#         """RERA"""

        try:
            rera = item.find('div', attrs={"class": "css-1lqxvi5"}).text
        except:
            rera = 'NOT RERA'
#         """detail"""

        try:
            detail = item.find('div',attrs={"class":"css-nqm4eq"}).text
        except:
            detail='NA'


        house_details = { 'prices': price, 'home_details': BHK,'prop_area':area,'emi':EMI,'Link':link,
                       'Rera':rera,'Detail':detail,'Owner':owner}
        # 'area': area
        list_data.append(house_details)
        
        
    print(f'page = {i}')
    time.sleep(1)
all_data = pd.DataFrame(list_data)
all_data.to_csv('rental.csv') 
```
